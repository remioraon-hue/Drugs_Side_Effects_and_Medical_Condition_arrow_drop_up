# **DATAFRAME**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
# The file is named '/content/drugs_side_effects_drugs_com.csv'
try:
    df = pd.read_csv('/content/drugs_side_effects_drugs_com.csv')
    print("CSV file loaded successfully!")
except FileNotFoundError:
    print("Error: The file 'drugs_side_effects_drugs_com.csv' was not found. Please ensure it is in the same directory as this script.")
    exit()
```

# **ANALYZING AND VISUALIZING SIDE EFFECTS**

```python
# --- Part 1: Analyzing and Visualizing Side Effects ---

# Split the 'side_effects' column to count individual side effects
# The side effects are listed as a single string
# We'll need to parse this string to get a count of each effect
all_side_effects = []
for effects_list in df['side_effects'].dropna():
    # Clean up the string by removing parentheses and extra spaces
    cleaned_effects = effects_list.replace('(', '').replace(')', '')
    # Split the string by comma and trim whitespace
    individual_effects = [effect.strip() for effect in cleaned_effects.split(',')]
    all_side_effects.extend(individual_effects)

# Create a pandas Series from the list of all side effects and get their counts
side_effects_counts = pd.Series(all_side_effects).value_counts()

# Let's plot the top 10 most common side effects
top_10_side_effects = side_effects_counts.head(10)

plt.figure(figsize=(12, 8))
top_10_side_effects.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Common Side Effects')
plt.xlabel('Side Effect')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# **Analyzing and Visualizing Drug Classes**

```python
# --- Part 2: Analyzing and Visualizing Drug Classes ---

# Similar to side effects, we will parse the 'drug_classes' column
all_drug_classes = []
for classes_list in df['drug_classes'].dropna():
    # Split the string by comma and trim whitespace
    individual_classes = [drug_class.strip() for drug_class in classes_list.split(',')]
    all_drug_classes.extend(individual_classes)

# Create a pandas Series from the list of all drug classes and get their counts
drug_classes_counts = pd.Series(all_drug_classes).value_counts()

# Plot the top 10 most common drug classes
top_10_drug_classes = drug_classes_counts.head(10)

plt.figure(figsize=(12, 8))
top_10_drug_classes.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Most Common Drug Classes')
plt.xlabel('Drug Class')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# **Additional Analysis - Most Common Medical Conditions**

```python
# --- Part 3: Additional Analysis - Most Common Medical Conditions ---

# Let's also analyze the 'medical_condition' column
medical_condition_counts = df['medical_condition'].value_counts()

# Plot the top 10 most common medical conditions
top_10_medical_conditions = medical_condition_counts.head(10)

plt.figure(figsize=(12, 8))
top_10_medical_conditions.plot(kind='bar', color='coral')
plt.title('Top 10 Most Common Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nAnalysis complete. Check the plots for visualizations of side effects, drug classes, and medical conditions.")
```
